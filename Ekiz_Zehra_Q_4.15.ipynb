Question 4.15=
 Assume a Gaussian stimulus distribution p(s) with mean 20 and standard deviation
4; this also serves as the prior distribution. The measurement distribution p(x|s) is Gaussian with
standard deviation Ïƒ = 5.


a. Draw 1000 values of s from the stimulus distribution. For each value of s, draw a single x
from the measurement distribution p(x|s).


In each question I labelled the y axes as "s values" while labelling y axes as estimation or x values.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm

sample_size = 1000

mu1 = 20
sigma1 = 4
sigma2 = 5

s_data = norm.rvs(loc = mu1, scale = sigma1, size = sample_size)
x_data = norm.rvs(loc = s_data, scale = sigma2, size = sample_size)

plt.figure()
plt.plot(s_data, 'ko', markersize = 2)
plt.title("s values")
plt.xlabel("samples")
plt.ylabel("s values")

plt.figure()
plt.plot(x_data, 'ko', markersize = 2)
plt.title("x values")
plt.xlabel("samples")
plt.ylabel("x values")
```

b. Scatter in a scatterplot the measurement x against the true stimulus s. In a separate scatterplot,
scatter the posterior mean estimate against the true stimulus. Set your axes suitably. Draw
the diagonal as a dashed black line. Draw a horizontal dashed black line at the prior mean.


```python
mean_posterior = (1/(sigma2**2)*s_data+mu1/(sigma1**2))/(1/sigma1**2+1/sigma2**2)
sigma_posterior = math.sqrt((1/sigma2**2)/((1/sigma1**2+1/sigma2**2)**2))
pme = norm.rvs(loc = mean_posterior, scale = sigma_posterior, size = sample_size)

a, b = np.polyfit(pme, s_data, 1)

plt.figure()
plt.scatter(x_data,s_data)
plt.axhline(y = 20, color = 'k', linestyle = '--')
plt.plot([min(x_data), max(x_data)], [min(s_data), max(s_data)], color = 'k', linestyle = '--')
plt.ylabel("s values")
plt.xlabel("x values")
plt.title("s vs x with sigma 5")

plt.figure()
plt.scatter(pme,s_data)
plt.axhline(y = 20, color = 'k', linestyle = '--')
plt.plot([min(pme), max(pme)], [min(s_data), max(s_data)], color = 'k', linestyle = '--')
plt.plot(pme, a*pme+b, color = 'r')
plt.ylabel("s values")
plt.xlabel("pme values")
plt.title("posterior sigma 5")
```

c. Repeat (a) and (b) using Ïƒ = 1 instead of Ïƒ = 5. The two scatterplots should look very
similar. Explain.


```python
mu1 = 20
sigma1 = 4
sigma2 = 1

s_data = norm.rvs(loc = mu1, scale = sigma1, size = sample_size)
x_data = norm.rvs(loc = s_data, scale = sigma2, size = sample_size)

mean_posterior = (1/(sigma2**2)*s_data+mu1/(sigma1**2))/(1/sigma1**2+1/sigma2**2)
sigma_posterior = math.sqrt((1/sigma2**2)/((1/sigma1**2+1/sigma2**2)**2))
pme = norm.rvs(loc = mean_posterior, scale = sigma_posterior, size = sample_size)

a, b = np.polyfit(pme, s_data, 1)

plt.figure()
plt.scatter(x_data,s_data)
plt.axhline(y = 20, color = 'k', linestyle = '--')
plt.plot([min(x_data), max(x_data)], [min(s_data), max(s_data)], color = 'k', linestyle = '--')
plt.ylabel("s values")
plt.xlabel("x values")
plt.title("s vs x with sigma 1")

plt.figure()
plt.scatter(pme,s_data)
plt.axhline(y = 20, color = 'k', linestyle = '--')
plt.plot([min(pme), max(pme)], [min(s_data), max(s_data)], color = 'k', linestyle = '--')
plt.plot(pme, a*pme+b, color = 'r')
plt.ylabel("s values")
plt.xlabel("pme values")
plt.title("posterior sigma 1")
```

These two scatterplots look so similar because our standard deviation change from our initial st.dev is really low. In the first estimation MLE, while the given st.dev. is 1, the posterior mean estimation's st.dev. is 16/17 so just like 1. Therefore, since the variances did not change the plots are similar.
